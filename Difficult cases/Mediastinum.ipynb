```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import matplotlib.image as mpimg
import re
import os 
from sklearn.metrics import roc_auc_score
```

```python
df = pd.read_csv('/home/storage/submarine-data/praram9/annotation/v2_30000_w_report/annotation.csv')
```

# pattern
mediastinal_pattern = 'mediastin|Mediastin'
No_pattern = '(nremarkable|No|ormal|no).{0,20}(mediastin|Mediastin)|(mediastin|Mediastin).{0,20}(ormal|nremarkable)|Heart and mediastinum'
image_path = '/home/storage/submarine-data/praram9/preprocess/ori_size/'
for i,j in df[['Image Index','Reports']].iterrows():
    if re.search(No_pattern,j[1]):
        j[1] = re.sub(No_pattern,'',j[1])
    if re.search(mediastinal_pattern,j[1]):
        image = mpimg.imread(image_path + j[0])
        plt.figure(figsize=(20,14))
        plt.imshow(image,cmap=plt.cm.bone)
        plt.show()
        # print index and roport 
        print(j[0])
        print(j[1])
