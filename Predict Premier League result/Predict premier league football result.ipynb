# Predict the result of match from halftime result using logistic regression 
Create X_train data 

```python
test = lambda x : True if (x>10 and x<20 ) else False
```

```python
test(15)
```

```python
df['Homeresult'] = df['FTR'].apply(lambda x : 2 if x =='H' else (1 if x == 'D' else 0) )
```

```python
df[['Homeresult','FTR']]
```

```python
# Assume that the number of yellow card and number of shot is estimate to be half on the half time 
df[['HY','AY','HR','AR','HS','AS','HST','AST']]= df[['HY','AY','HR','AR','HS','AS','HST','AST']]/2
```

```python
from tensorflow.keras.losses import SparseCategoricalCrossentropy
X = df[['HTHG','HTAG','HY','AY','HR','AR','HS','AS','HST','AST']].values
y = df[['Homeresult']].values
model = Sequential([Dense(units = 10, activation = 'linear'),
                    Dense(units = 3,activation = 'softmax')])
model.compile(loss = SparseCategoricalCrossentropy(),optimizer = tf.keras.optimizers.Adam(0.001))
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2)
model.fit(X_train,y_train,epochs = 80)

```

```python
# 1st column mean the chance for homeclub to lose the match  (label 0)
# 2nd column mean the chance for homeclub to draw the match (label1)
# 3rd column mean the chance for homeclub to win the macth (label2)
for i in model.predict(X_test) :
    newlist = ['{:.4f}'.format(x) for x in list(i)]
    print(newlist)
    
```

```python
predict = model.predict(X_test)
```

```python
# calculate loss functioin for model by using y_test / predict 
from math import log
loss = 0
for i in range(len(y_test)) :
    index = y_test[i][0]
    result = predict[i,index] 
    loss += -log(result)
    loss = loss/(len(y_test))
loss
```

```python
# loss function in traning set
from math import log
loss = 0
train_predict = model.predict(X_train)
for i in range(len(y_train)) :
    index = y_train[i][0]
    result = train_predict[i,index] 
    loss += -log(result)
    loss = loss/(len(y_train))
'{:.8f}'.format(loss)
```

```python
# calculate loss of soft max = -log(probability of being A) if the y_test actually is A 
x = np.linspace(0.01,1.2)
y = [-log(i) for i in x ]
plt.plot(x,y,color ='b')
```

```python
# accuracy of model in test set 
TP = 0
for i in range(len(y_test)) :
    index = y_test[i][0]
    result = predict[i,index]
    # set threshold to be 0.5
    if result >= 0.5 :
        TP+=1
TP/len(y_test)

# when AI predict as win lose or draw the chance that the prediction will be correct is 46%
```
